```python
data_path = "data"
# find all files in the data directory
files = os.listdir(data_path)
files = sorted(files)
count = 0
stressed = []
not_stressed = []

# x is a pd dataframe with the features

# gt is a list of booleans, true if stressed, false if not
accuracies=[]
for t in tqdm(range(0,36)):
    unstressed_file = "Subject"+str(t).rjust(2, '0')+"_1.edf"
    stressed_file = "Subject"+str(t).rjust(2, '0')+"_2.edf"
    # read the edf file
    gt=[]
    x=[]
    is_stressed = False
    for file in [unstressed_file, stressed_file]:
        helper= EDFHelper("data/"+str(file), sampleRate=500, lowpass=30, highpass=0.5, windowSize=500)
        for eeg in helper:
            bands = eeg.bandPower()
            dat = []
            for channel in bands:
                t=[v for k, v in channel.items()]
                mean = np.mean(t)
                dat.extend([v/mean for k, v in channel.items()])
            x.append(dat)
            gt.append(is_stressed)
        is_stressed = True

    x=pd.DataFrame(x)
    # convert x from a list to a pd dataframe

    # convert gt from a list to a pd dataframe
    x_train, x_test, y_train, y_test = train_test_split(x, gt, test_size=0.3, random_state=42)
    # convert the dataframes to numpy arrays
    # create the vectorizer
    # fit the vectorizer to the training data
    # scale the training data

    # scale the testing data
    pca_30 = PCA(n_components=30, random_state=69)
    x_train = pca_30.fit_transform(x_train)
    x_test = pca_30.transform(x_test)
    # create the random forest classifier
    clf = RandomForestClassifier()
    # train the classifier
    clf.fit(x_train, y_train)
    # predict the test data
    y_pred = clf.predict(x_test)
    #print(y_pred)
    # print the accuracy
    accuracies.append(accuracy_score(y_test, y_pred))
    #print("Accuracy:", accuracy_score(y_test, y_pred))
    # print the confusion matrix
    #print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    # print the classification report
    #print("Classification Report:\n", classification_report(y_test, y_pred))

print(accuracies)
print("Average Accuracy:", np.mean(accuracies))
```
